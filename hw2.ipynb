```python
#Homework 11
def substitution(message,sub_alpha):
    ''' 
    Substitution Cipher
    Accepts 2 parameters:
    -a string message that we want to encode or decode
    -a list of 28 characters for the substitution alphabet as sub_alpha
    '''
    alpha_dict = {}
    alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ' "
    for a in alpha:
        if a not in alpha_dict:
            alpha_dict[a] = 0
    for char,sub_char in zip(alpha,sub_alpha):
        alpha_dict[char] = sub_char
    stri = ''
    for i in message:
        stri = stri + alpha_dict[i]
    return stri
```
